## Total de Vendas por Ano


```python
vendas_ano = carros.groupby(["Ano"]).sum()["ValorVenda"].rename("Total de Vendas por Ano")

analise_venda_ano = pd.DataFrame({"Total de Vendas por Ano" : vendas_ano})

analise_venda_ano["Total de Vendas por Ano"] = analise_venda_ano["Total de Vendas por Ano"].map("R$ {:,.2f}".format) 

analise_venda_ano
```

## Custo de Mão de Obra Por Estado 

```python
custo_entrega_fabricante = carros.groupby(["Fabricante"]).sum()["CustoEntrega"].rename("Custo de Entrega por Fabricante")

analise_custo_fabricante = pd.DataFrame({"Custo de Entrega por Fabricante" : custo_entrega_fabricante})

analise_custo_fabricante["Custo de Entrega por Fabricante"] = analise_custo_fabricante["Custo de Entrega por Fabricante"].map("R$ {:,.2f}".format)

analise_custo_fabricante
```

## Custo de Mão de Obra Por Estado 

```python
custo_mao_obra_estado = carros.groupby(["Estado"]).sum()["CustoMaoDeObra"].rename("Custo Mão de Obra por Estado") 

analise_custo_mao_obra_estado = pd.DataFrame({"Custo Mão de Obra por Estado" : custo_mao_obra_estado})

analise_custo_mao_obra_estado["Custo Mão de Obra por Estado"] = analise_custo_mao_obra_estado["Custo Mão de Obra por Estado"].map("R$ {:,.2f}".format) 

analise_custo_mao_obra_estado
```

## Total de Vendas Geral

```python
vendas_geral_total = carros["ValorVenda"].sum()

analise_geral_total_vendas = pd.DataFrame({"Total de Vendas" : [vendas_geral_total]})

analise_geral_total_vendas["Total de Vendas"] = analise_geral_total_vendas["Total de Vendas"].map("R$ {:,.2f}".format) 
analise_geral_total_vendas
```

## Total de Vendas por Estado 

```python
vendas_estado_total = carros.groupby(["Estado"]).sum()["ValorVenda"].rename("Total de Vendas por Estado") 
vendas_estado_media = carros.groupby(["Estado"]).mean()["ValorVenda"].rename("Média de Vendas por Estado")

analise_estado_vendas_geral = pd.DataFrame({"Total de Vendas por Estado" : vendas_estado_total, 
                                     "Média de Vendas por Estado" : vendas_estado_media})

analise_estado_vendas_geral["Total de Vendas por Estado"] = analise_estado_vendas_geral["Total de Vendas por Estado"].map("R$ {:,.2f}".format)
analise_estado_vendas_geral["Média de Vendas por Estado"] = analise_estado_vendas_geral["Média de Vendas por Estado"].map("R$ {:,.2f}".format)

analise_estado_vendas_geral
```

## Total de Vendas por Fabricante, Estado e Ano

```python
vendas_fabricante_estado_ano_total = carros.groupby(["Fabricante", "Ano", "Estado"]).sum()["ValorVenda"].rename("Total de Vendas Geral")
vendas_custos_fabricante_estado_ano_total = carros.groupby(["Fabricante", "Ano", "Estado"]).sum()["ValorCusto"].rename("Total Vendas Custo")
custos_entrega_fabricante_estado_ano_total = carros.groupby(["Fabricante", "Ano", "Estado"]).sum()["CustoEntrega"].rename("Total de Custo Entrega")
custos_mão_de_obra_fabricante_estado_ano_total = carros.groupby(["Fabricante", "Ano", "Estado"]).sum()["CustoMaoDeObra"].rename("Total de Custo Mão de Obra")

analise_geral_f_e_a = pd.DataFrame({"Total de Vendas Geral" : vendas_fabricante_estado_ano_total, 
                                    "Total Vendas Custo" : vendas_custos_fabricante_estado_ano_total,
                                    "Total de Custo Entrega" : custos_entrega_fabricante_estado_ano_total,
                                    "Total de Custo Mão de Obra" : custos_mão_de_obra_fabricante_estado_ano_total}) 

analise_geral_f_e_a["Total de Vendas Geral"] = analise_geral_f_e_a["Total de Vendas Geral"].map("R$ {:,.2f}".format)
analise_geral_f_e_a["Total Vendas Custo"] = analise_geral_f_e_a["Total Vendas Custo"].map("R$ {:,.2f}".format)
analise_geral_f_e_a["Total de Custo Entrega"] = analise_geral_f_e_a["Total de Custo Entrega"].map("R$ {:,.2f}".format)
analise_geral_f_e_a["Total de Custo Mão de Obra"] = analise_geral_f_e_a["Total de Custo Mão de Obra"].map("R$ {:,.2f}".format)

analise_geral_f_e_a.head(60)
```

## Gráfico de Modelos por Valores de Venda, Valor de Custo, Custo de Entrega, Custo Mão de Obra 

```python
COLUMNS = ["ValorVenda", "ValorCusto" , "CustoEntrega", "CustoMaoDeObra"] 
COLORS = ["#386cb0", "#fdb462", "#7fc97f"] 
MODELOS = ["Arnage", "Azure", "Camargue", "Cerbera", "DB4", "DB7", "DB9", "DBS", "GT", "Phantom", 
           "Rapide", "TR4", "TR5", "TR7", "Turbo R", "Tuscan", "Vanquish", "Vantage", "Wraith", "XJ12"
           "XJ6", "XK", "Zagato"]

fig, axes = plt.subplots(4, 4, figsize = (12, 8), sharex="col", tight_layout=True)

for i in range(len(COLUMNS)):
    
    for j in range(len(COLUMNS)): 
        
        if i > j: 
            
            for modelos, color in zip(MODELOS, COLORS):
                
                data = carros[carros["Modelo"] == modelos] 
                axes[i, j].scatter(COLUMNS[j], COLUMNS[i], color=color, alpha=0.5, data=data)
                
        if i == j: 
            
            for modelos, color in zip(MODELOS, COLORS): 
                
                data = carros[carros["Modelo"] == modelos] 
                axes[i,j].hist(COLUMNS[j], bins=15, alpha=0.5, data=data)
```

## Removendo Figuras vazias no Gráfico de Modelos

```python
for i in range(len(COLUMNS)): 
    
    for j in range(len(COLUMNS)):
        
        if i < j: 
            
            axes[i, j].remove()
            
fig
```
